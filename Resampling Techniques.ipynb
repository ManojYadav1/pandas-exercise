## 4. Data Preprocessing

```python
sales_squeezed_data = pd.read_csv('sales.csv', index_col = 'Month', parse_dates = True ,squeeze = True)
sales_squeezed_data
```

```python
type(sales_squeezed_data)
```

```python
resampled_data = sales_squeezed_data.resample('D').mean()
resampled_data.head(40)
```

### 4.1 UpSampling using Linear Interpolation

```python
interpolated_upsampling = resampled_data.interpolate(method = 'linear')
interpolated_upsampling.head(40)
```

### 4.2 Check with Original data using Visuals

#### After Interpolation

```python
from matplotlib import pyplot as plt

plt.figure(figsize = (16,8))
interpolated_upsampling.plot()
plt.title('Shampoo sales - After Interpolation',size = 30)
plt.show()
```

####  Before Interpolation

```python
plt.figure(figsize = (16,8))
sales_squeezed_data.plot()
plt.title('Shampoo sales - Before Interpolation',size = 30)
plt.show()
```

### Downsampling

```python
downsampling_quarter_wise = interpolated_upsampling.resample('Q').mean()
downsampling_quarter_wise
```

```python
plt.figure(figsize = (16,8))
downsampling_quarter_wise.plot()
plt.title('Shampoo quarterly sales',size = 30)
plt.show()
```
